```python
# how many petaints age are greter than 47.
data_filled[data_filled['age']>47].shape[0]
```

```python
#value Counts....How many male and female in the data.
data_filled['gender'].value_counts() #How many male and female in the data.
data_filled['region'].value_counts()  #How many region in the data.
```

```python
#Maximum 10 claim value with use sorting
data_filled.sort_values(by='claim',ascending=False)[:10] 
```

```python
#Filter the data with Column gender,smoker,claim.
data_filled[(data_filled['gender']=='female') & (data_filled['smoker']=='Yes') & (data_filled['claim']>50000)]
```

```python
# Filter the data with specific condition.
data_filled[data_filled['age'].between(18,25)]['claim'].sum()
```

```python
# Visualize the Histograme plot of Age.
plt.figure(figsize=(10, 6))
sns.histplot(data_filled['claim'], kde=True)
plt.title('Histograme plot of Claim')
plt.xlabel('claim')
plt.ylabel('Frequency')
plt.show()
```

```python
# Visualize the relationship between Age And Claim using Scatter Plot.
plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='claim', data=data_filled)
plt.title('Relationship between Age and Claim')
plt.xlabel('age')
plt.ylabel('claim')
plt.show()
```

```python
# Visualize the count of a Gender Using Count plot.
plt.figure(figsize=(10, 6))
sns.countplot(x='gender', data=data_filled)
plt.title('Count of Gender')
plt.xlabel('gender')
plt.ylabel('Count')
plt.show()
```
